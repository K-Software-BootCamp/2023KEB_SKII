# plot

```python
def plot_data(data, bearing_name, col, color, plot_type=None):
    # plot_type = 1 : Horiz, Vert col plot
    # plot_type = 2 : two data plot
    base_dir = '/content/drive/MyDrive/plots'

    # Create a figure and axis
    fig, ax = plt.subplots()

    if plot_type==None:
        # Reset the index of the DataFrame
        data_reset = data.reset_index(drop=True)

        # Plot the data using integer index as x-axis
        ax.plot(data_reset.index, data[col], color=color)

        # Customize the plot
        ax.set_xlabel('Time Index')
        ax.set_ylabel('Amplitude')
        ax.set_title(f'{col} amplitude for {bearing_name}')

        # save plot
        plt.savefig(f'{base_dir}/{col} amplitude for {bearing_name}')

    elif plot_type==1:
        # Reset the index of the DataFrame
        data_reset = data.reset_index(drop=True)

        # Plot the data using integer index as x-axis
        ax.plot(data_reset.index, data[col[0]], color=color[0], label=f'Amplitude ({col[0]})')
        ax.plot(data_reset.index, data[col[1]], color=color[1], label=f'Amplitude ({col[1]})')

        # Customize the plot
        ax.set_xlabel('Time Index')
        ax.set_ylabel('Amplitude')
        ax.set_title(f'{col[0]} & {col[1]} amplitude for {bearing_name}')

        # Add a legend
        ax.legend()

        # save plot
        plt.savefig(f'{base_dir}/{col[0]} & {col[1]} amplitude for {bearing_name}')

    elif plot_type==2:
        # Plotting the first dataset
        plt.plot(data[0].index, data[0][col], color=color[0], label='Full Test Set')
        # Plotting the second dataset
        plt.plot(data[1].index, data[1][col], color=color[1], label='Test Set')

        # Customize the plot
        ax.set_xlabel('Time Index')
        ax.set_ylabel('Amplitude')
        ax.set_title(f'{col} amplitude for {bearing_name}')

        # Add a legend
        ax.legend()

        # save plot
        plt.savefig(f'{base_dir}/{col} amplitude for {bearing_name}')

    # Show the plot
    plt.show()
```

```python
def Learning_plot(data, bearing_name):
    plot_data(data, bearing_name, 'Horiz', 'dodgerblue')
    plot_data(data, bearing_name, 'Vert', 'orange')
    plot_data(data, bearing_name, ['Horiz', 'Vert'], ['dodgerblue', 'orange'], 1)
```

```python
def Test_plot(data, bearing_name):
    plot_data(data[0], bearing_name+'(Full Test)', 'Horiz', 'dodgerblue')
    plot_data(data[0], bearing_name+'(Full Test)', 'Vert', 'orange')
    plot_data(data, bearing_name+'(Full Test & Test)', 'Horiz', ['dodgerblue', 'deeppink'], plot_type=2)
    plot_data(data, bearing_name+'(Full Test & Test)', 'Vert', ['orange', 'deeppink'], plot_type=2)
```
