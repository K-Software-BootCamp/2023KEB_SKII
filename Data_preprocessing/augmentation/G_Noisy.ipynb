```python
import pandas as pd
import numpy as np
import os
import zipfile

from datetime import datetime, time
```

```python
# CSV 파일들이 있는 경로
csv_folder_path = 'D:/data/Learning_set/Bearing3_1/'
```

```python
noisy_folder_path_bearing1_1 = 'D:/data/noisy015b3_1'
```

```python
# List all files in the folder
csv_files = [f for f in os.listdir(csv_folder_path) if f.endswith('.csv') and 'acc' in f]

# Sort the file names
csv_files.sort()

# Load and process each CSV file
for csv_file in csv_files:
    file_path = os.path.join(csv_folder_path, csv_file)
    df = pd.read_csv(file_path, header=None)  # Load the CSV file without specifying column names
    print(f"Loaded {csv_file}, shape: {df.shape}")

    # Apply noise to the 'Horiz' and 'Vert' columns
    noise_factor = 0.15  # Adjust the noise factor as needed
    df[4] = df[4] + np.random.normal(0, noise_factor, len(df))
    df[5] = df[5] + np.random.normal(0, noise_factor, len(df))

    # 새로운 파일로 저장할 수도 있습니다
    new_file_path = os.path.join(noisy_folder_path_bearing1_1, f"noisy_0.15{csv_file}")
    df.to_csv(new_file_path, index=False, header=False)  # Save without header

    print(f"Noisy data added to {csv_file}, shape: {df.shape}")
```
