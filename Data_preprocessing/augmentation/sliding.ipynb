```python
import pandas as pd

# 슬라이딩 기법에 쓸 함수 구현
def sliding_window_augmentation(df, window_size):
    augmented_data = []
    num_rows = df.shape[0]

    for start_idx in range(0, num_rows - window_size + 1):
        end_idx = start_idx + window_size
        window_data = df.iloc[start_idx:end_idx].copy()
        augmented_data.append(window_data)

    return pd.concat(augmented_data, ignore_index=True)
```
